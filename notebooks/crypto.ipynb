```python
def browser_table(df):
    import webbrowser
    import os

    head = '''
    <!DOCTYPE html><html>
    <head>
    <title>demo</title>
    <link href="style.css" rel="stylesheet">
    <script src="sorttable.js"></script>
    </head>
    <body>
    '''
    table = df.to_html()
    table = table.replace('class="dataframe"', 'class="sortable"')

    foot = '''</body></html>'''


    path = os.path.abspath('temp.html')
    url = 'file://' + path

    with open(path, 'w') as f:
        f.write(head + table + foot)
    webbrowser.open(url)
browser_table(df)
```
